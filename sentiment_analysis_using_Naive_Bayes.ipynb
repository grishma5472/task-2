```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('sentiment_analysis_dataset.csv')
data.head()
```

```python
data.shape
```

```python
#Sentiments to classify into are Angry, Happy, Neutral
#Multinomial Naive Bayes
data.describe()
```

```python
# data['Angry']=data['sentiment'].apply(lambda x:1 if x=='Angry' else 0)
# data['Happy']=data['sentiment'].apply(lambda x:1 if x=='Happy' else 0)
# data['Neutral']=data['sentiment'].apply(lambda x:1 if x=='Neutral' else 0)

#emotion: 0 is angry, 1 is happy, 2 is neutral
le= LabelEncoder()
data['emotion']=le.fit_transform(data['sentiment'])
data[0:10]
```

```python
data[10:20]
```

```python
data[20:30]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data['line'], data['emotion'], test_size=0.3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
x_train_count=v.fit_transform(x_train.values)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_count,y_train)
```

Results on target text

```python
statements=[
    #angry
    'I am really annoyed by your constant complaining and you never offer any solutions which is very unhelpful and negative',
    'It is frustrating that you never pay attention during discussions and your lack of focus is really affecting our progress',
    #happy
    'I am delighted by your friendliness and you always make everyone feel welcome which fosters a sense of community',
     'It is wonderful that you always show kindness and your empathy towards others is truly heartwarming and appreciated',
   #neutral
    'Your analysis of the data was accurate and well presented providing a clear understanding of the trends and patterns',
    'The meeting minutes you prepared were detailed and well organized accurately reflecting the discussions and decisions made'
]

#predictions
model.predict(v.transform(statements))
```

```python
x_test_count=v.transform(x_test)
score=model.score(x_test_count, y_test)
print(f'Accuracy score: {score}')
```
